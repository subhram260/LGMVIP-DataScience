<h2>Import the modules</h2>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as pe
pd.set_option("display.max_columns",None)
pd.get_option("display.max_columns")
```

<h2>Load Dataset

```python
df = pd.read_csv('globalterrorismdb_0718dist.csv',encoding='latin1')

df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

<h2>Find all Columns name

```python
pd.DataFrame(columns=df.columns.values,index=None)
```

<h2>Give appropriate names

```python
df.rename(columns={'iyear':'YEAR','imonth':'MONTH','iday':'DAY','extended':'EXTENDED','country_txt':'COUNTRY',
                   'success':'SUCCESS','region_txt':'REGION','provstate':'STATE','city':'CITY','gname':'GROUP',
                   'attacktype1_txt':'ATTACK TYPE','latitude':'LATITUDE','longitude':'LONGITUDE',
                   'nkill':'KILLED','nwound':'WOUNDED','target1':'TARGET'},inplace=True)
```

<h2>Filtering useful Columns

```python
df=df[['YEAR','MONTH','DAY','EXTENDED','COUNTRY','REGION','STATE','CITY','LATITUDE',
       'LONGITUDE','GROUP','ATTACK TYPE','SUCCESS','KILLED','WOUNDED','TARGET']]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.hist(figsize=(25,15))
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(np.round(df.corr(),2),annot=True)
```

<h4>  From the above Heatmap we came to know that Number of Wounded is Highly related to Number of killed 

```python
fig,ax = plt.subplots(figsize=(20,7))
sns.countplot(x='YEAR',data=df,palette='viridis')
plt.xticks(rotation=90)
plt.title("Number of Attacks every year")
plt.ylabel("Number of attack")
plt.xlabel("\nYEAR")
```

<h4>  From the above graph we can find out the number of attacks in each year. And we came to know that there were a large number of attacks in
<br>
<br>
     2014.

```python
fig,ax = plt.subplots(figsize=(20,7))
sns.barplot(x=df['REGION'].value_counts().index,y=df['REGION'].value_counts().values,palette='viridis')
plt.xticks(rotation=90)
plt.title("Number of Attacks in Each Region")
plt.ylabel("Number of attack")
plt.xlabel("\nREGION")
```

<h4>  From the above graph we came to know that there were a large number of attacks in Middle East & North Africa.

```python
fig,ax = plt.subplots(figsize=(20,7))
sns.barplot(x=df['CITY'].value_counts()[1:50].index,y=df['CITY'].value_counts()[1:50].values,palette='viridis')
plt.xticks(rotation=90)
plt.title("Number of Attacks in Each City")
plt.ylabel("Number of attack")
plt.xlabel("\nCITY")
```

```python
fig,ax = plt.subplots(figsize=(20,7))
sns.barplot(x=df['GROUP'].value_counts()[1:50].index,y=df['GROUP'].value_counts()[1:50].values,palette='viridis')
plt.xticks(rotation=90)
plt.title("Number of Attacks by Each Groups")
plt.ylabel("Number of attack")
plt.xlabel("\nGROUP")
```

```python
fig,ax = plt.subplots(figsize=(20,7))
sns.barplot(x=df['ATTACK TYPE'].value_counts().index,y=df['ATTACK TYPE'].value_counts().values,palette='viridis')
plt.xticks(rotation=90)
plt.title("Number of Attacks by Attack Type")
plt.ylabel("Number of attack")
plt.xlabel("\nATTACK TYPE")
```

```python
fig,ax = plt.subplots(figsize=(20,7))
sns.barplot(x=df['TARGET'].value_counts()[:50].index,y=df['TARGET'].value_counts()[:50].values,palette='viridis')
plt.xticks(rotation=90)
plt.title("Number of Attacks based on Targets")
plt.ylabel("Number of attack")
plt.xlabel("\nTARGETS")
```

```python
print("All = Total number of Attacks")
print("1 = Total number of Successful Attacks")
print("0 = Total number of Unsuccessful Attacks")
pd.crosstab([df.COUNTRY],[df.SUCCESS],margins=True).style.set_properties(**{'background-color': 'black','color': 'lawngreen',})
```

```python
te=df.groupby(['COUNTRY'],as_index=False).count()
fig = pe.choropleth(te,locations='COUNTRY',locationmode='country names',color='YEAR',
                    hover_name='COUNTRY',projection='orthographic',title="Total Numbers of attacks",
                   labels={'COUNTRY':'Country','YEAR':'Attack'})
fig.show()
```
